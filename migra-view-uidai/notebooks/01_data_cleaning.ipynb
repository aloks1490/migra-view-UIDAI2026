```python
import pandas as pd
```

```python
# Load the datasets  
df_bio = pd.read_csv('/Users/alokranjansingh/Desktop/uidaiiiii/migra-view-uidai/data/raw/combined_aadhar_biometric.csv')
df_enrol = pd.read_csv('/Users/alokranjansingh/Desktop/uidaiiiii/migra-view-uidai/data/raw/combined_aadhar_enrolment.csv')
df_demo = pd.read_csv('/Users/alokranjansingh/Desktop/uidaiiiii/migra-view-uidai/data/raw/combined_aadhar_demographic.csv')
```

```python
# 1. Clean Dates: Convert to datetime objects for time-series analysis [cite: 13, 14, 16]
for df in [df_enrol, df_demo, df_bio]:
    df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
```

```python
# 2. Merge Data: Join on common columns (date, state, district, pincode) [cite: 13, 14, 16]
df_master = pd.merge(df_enrol, df_demo, on=['date', 'state', 'district', 'pincode'], how='outer')
df_master = pd.merge(df_master, df_bio, on=['date', 'state', 'district', 'pincode'], how='outer')
```

```python
# 3. Handle Nulls: Fill missing numeric values with 0
cols_to_fill = ['age_0_5', 'age_5_17', 'age_18_greater', 'demo_age_5_17', 'demo_age_17_', 'bio_age_5_17', 'bio_age_17_']
df_master[cols_to_fill] = df_master[cols_to_fill].fillna(0)
```

```python
# Save for next notebook
df_master.to_csv('/Users/alokranjansingh/Desktop/uidaiiiii/migra-view-uidai/data/processed/processed_aadhaar_data.csv', index=False)
print("Data Cleaning Complete. Master file saved.")
```
